```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df.head(2)
```

```python
from typing_extensions import final
country_names=final_df.Country.value_counts().index
country_val=final_df.Country.value_counts()
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observations:
Most of orders of zomato are in India

Second is usa

Third is Uk


```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Counts'})
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Counts',data=ratings,hue='Rating color',palette=['white','red','orange','yellow','green','green'])
```

Observation:Avg rating are between 2.5 to 3.5


```python
final_df[final_df['Rating color']=='Dark Green'].groupby('Country').size().reset_index()
```

Observations:Max number of Ratings above 4.5 are from India
